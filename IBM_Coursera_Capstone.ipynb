#### Getting site data and scrapping with BeatifulSoup

```python
soup=BeautifulSoup(web_pg.content, 'lxml')
my_table=soup.findAll(id='maincontent')
#print(my_table)
#print(len(my_table))
```

```python
list_code_n=[]
list_names_n=[]
for item in my_table:
    for its in item.find_all('dl'):
        x=its.find_all('dt')
        y=its.find_all('dd')
        list_code_n.append(x)
        list_names_n.append(y)
```

```python
#print(list_code)
list_code=[]
list_names=[]
for row in list_code_n:
    for pcode in row:
        list_code.append(pcode.text)
```

```python
for row_1 in list_names_n:
    for pname in row_1:
        list_names.append(pname.text)
```

### Exploring all the neighborhoods returned by Fooursquare

```python
def getVenues(names, latitudes, longitudes, radius=500):
    
    list_of_venues=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url_1 = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            limit)
            
        # make the GET request
        results_1 = requests.get(url_1).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        list_of_venues.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results_1])

    nearby_venues_1 = pd.DataFrame([item for venue_list in list_of_venues for item in venue_list])
    nearby_venues_1.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues_1)
```

```python
#code that runs the above function on each neighborhood and creates a new dataframe by the name Toronto_venues_1
london_venues_1 = getVenues(names=final_df['postcode_name'],
                                   latitudes=final_df['latitude'],
                                   longitudes=final_df['longitude']
                                  )
```

#### Analyzing each neighborhood

```python
# one hot encoding
london_onehot = pd.get_dummies(london_venues_1[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
london_onehot['Neighborhood'] = london_venues_1['Neighborhood'] 

# move neighborhood column to the first column
new_columns = [london_onehot.columns[-1]] + list(london_onehot.columns[:-1])
london_onehot = london_onehot[new_columns]

london_onehot.head()
```

```python
london_onehot.shape
```

```python
#grouping the rows by neighborhoods and the mean of the frequency of the occurence of each category
london_df_grouped = london_onehot.groupby('Neighborhood').mean().reset_index()
london_df_grouped.head()
```

```python
#size of the new dataframe
london_df_grouped.shape
```

#### Getting the neighborhood along with the top 5 most common venues

```python
#Getting the neighborhoods along with the top 5 most common venues
num_of_venues = 5

for hood in london_df_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp_venue = london_df_grouped[london_df_grouped['Neighborhood'] == hood].T.reset_index()
    temp_venue.columns = ['venue','freq']
    temp_venue = temp_venue.iloc[1:]
    temp_venue['freq'] = temp_venue['freq'].astype(float)
    temp_venue = temp_venue.round({'freq': 3})
    print(temp_venue.sort_values('freq', ascending=False).reset_index(drop=True).head(num_of_venues))
    print('\n')
```

```python
#A function that returns the venues in a descending order
def common_venues(row, num_common_venues):
    row_categories = row.iloc[1:]
    row_categories_desc = row_categories.sort_values(ascending=False)
    
    return row_categories_desc.index.values[0:num_common_venues]
```

#### Dataframe that gives the top 10 venues for each neighborhood

```python
#A dataframe that Gives the top 10 venues for each neighborhood
num_common_venues = 10

premise_indicator = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for indicator in np.arange(num_common_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(indicator+1, premise_indicator[indicator]))
    except:
        columns.append('{}th Most Common Venue'.format(indicator+1))

# create a new dataframe
neighborhoods_venues_desc = pd.DataFrame(columns=columns)
neighborhoods_venues_desc['Neighborhood'] = london_df_grouped['Neighborhood']

for indicator in np.arange(london_df_grouped.shape[0]):
    neighborhoods_venues_desc.iloc[indicator, 1:] = common_venues(london_df_grouped.iloc[indicator, :], num_common_venues)

neighborhoods_venues_desc.head()
```

#### Clustering the neighborhoods with 5 clusters

```python
#initiating 5 clusters
# set number of clusters
kclusters = 5

london_grouped_clustering = london_df_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(london_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
#A Dataframe that has the cluster and the top 10 venues for each neighborhood
# add clustering labels
#neighborhoods_venues_desc.insert(0, 'Cluster Labels', kmeans.labels_)

london_merged = final_df
london_merged.rename(columns={'postcode_name':'Neighborhood'}, inplace=True)

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
london_merged = london_merged.join(neighborhoods_venues_desc.set_index('Neighborhood'), on='Neighborhood')

london_merged.head(10) # check the last columns!
```

```python
london_merged_final = london_merged.drop(london_merged.index[36]).reset_index(drop=True)
london_merged_final.head(10)
```

```python
london_merged_final['Cluster Labels'] = london_merged_final['Cluster Labels'].apply(np.int64)
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat_1, lon_1, poi, cluster in zip(london_merged_final['latitude'], london_merged_final['longitude'], london_merged_final['Neighborhood'], london_merged_final['Cluster Labels']):
    label = folium.Popup(poi + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat_1, lon_1],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
